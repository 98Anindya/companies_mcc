```python
import pandas as pd

# Load data from CSV file
data = pd.read_csv('tab_data.csv')

# Check for duplicated rows
duplicated_rows = data[data.duplicated(keep=False)]

if not duplicated_rows.empty:
    # Save duplicated rows to a new CSV file
    duplicated_rows.to_csv('duplicate_data.csv', index=False)
    
    # Display number of duplicate rows and their row numbers
    print(f"Number of duplicate rows found: {len(duplicated_rows)}")
    print("Duplicate rows found in rows:", ", ".join(map(str, duplicated_rows.index)))
else:
    print("No duplicate rows found.")
```

```python
import pandas as pd

# Load data from CSV file
data = pd.read_csv('Modified_Cleaned_1.csv')

# Find duplicated rows
duplicated_rows = data[data.duplicated(keep=False)]

if not duplicated_rows.empty:
    # Group by the entire row to find unique rows
    grouped = duplicated_rows.groupby(list(duplicated_rows.columns))
    
    # Iterate over unique rows
    for _, group in grouped:
        # Find the row numbers of the duplicates
        duplicate_indices = group.index.tolist()
        
        # Create a DataFrame containing original row and its duplicates
        duplicated_data = data.loc[duplicate_indices]
        
        # Write the duplicated data to a new CSV file
        duplicated_data.to_csv(f'duplicate_row_{duplicate_indices[0]}.csv', index=False)
        
        # Display information about duplicated rows
        print(f"Original Row Number: {duplicate_indices[0]}")
        print(f"Number of Duplicates: {len(duplicate_indices)}")
        print(f"Duplicated Row Numbers: {', '.join(map(str, duplicate_indices))}")
        print("-" * 50)
else:
    print("No duplicate rows found.")
```

```python
import pandas as pd

# Load data from CSV file
data = pd.read_csv('Modified_Cleaned_1.csv')

# Find duplicated rows
duplicated_rows = data[data.duplicated(keep=False)]

if not duplicated_rows.empty:
    # Iterate over each duplicated row
    for _, row in duplicated_rows.iterrows():
        # Find rows that are exactly identical (excluding the index column)
        identical_rows = data[(data == row).all(axis=1) & (data.index != row.name)]
        
        # If any identical rows are found, print information
        if not identical_rows.empty:
            print(f"Original Row Number: {row.name}")
            print(f"Number of Identical Duplicates: {len(identical_rows)}")
            print(f"Identical Duplicate Row Numbers: {', '.join(map(str, identical_rows.index))}")
            print("-" * 50)
else:
    print("No duplicate rows found.")
```
