```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('tab_data.csv')

# Display basic info and summary statistics
print(df.info())
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Exclude non-numeric columns for correlation matrix calculation
numeric_cols = df.select_dtypes(include=['number']).columns
corr_matrix = df[numeric_cols].corr()

# Plot a correlation matrix to understand relationships between features
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Plot histograms for each numerical feature to understand distributions
df[numeric_cols].hist(figsize=(15, 12), bins=20)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Load the data
df = pd.read_csv('tab_data.csv')

# Basic Statistics
print("Basic Statistics:")
print(df.describe())

# Data Distribution
print("\nData Distribution:")
df.hist(figsize=(10,10))
plt.show()

# Box plots for outlier detection
print("\nBox plots for outlier detection:")
df.boxplot(figsize=(25,25))
plt.show()

# Class Distribution
print("\nClass Distribution:")
print(df['clusterid - Prediction/OutPut'].value_counts())

# Correlation between features
print("\nCorrelation between features:")
# Exclude non-numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.show()

# Outlier detection using Z-score
print("\nOutlier detection using Z-score:")
z_scores = np.abs(stats.zscore(numeric_df))
threshold = 3
outliers = np.where(z_scores > threshold)
unique_outliers = np.unique(outliers[0])
print(f"Number of unique outliers: {len(unique_outliers)}")

# Display the rows with outliers
if len(unique_outliers) > 0:
    print("Rows with outliers based on Z-score:")
    print(df.iloc[unique_outliers])

# Save the cleaned data to a new CSV file
#cleaned_file_path = 'Cleaned_data_2.csv'
#df.to_csv(cleaned_file_path, index=False)
#print(f"\nCleaned data saved to {cleaned_file_path}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('tab_data.csv')

# Exclude the 'companycode' column from the correlation calculation
df_numeric = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix for numeric columns only
corr_matrix = df_numeric.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)

# Save the figure
plt.savefig('correlation_matrix_tab_data.jpeg', format='jpeg', dpi=300)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Cleaned_1.csv')  # Ensure this is the correct file name

# Exclude the 'companycode' column from the correlation calculation
df_numeric = df.select_dtypes(include=[np.number])
df_numeric = df_numeric.drop(columns=['companycode'], errors='ignore')  # Drop 'companycode' if it exists

# Calculate the correlation matrix
correlations = df_numeric.corr()

# Extract the correlations with the target variable and sort them
target_correlations = correlations['clusterid - Prediction/OutPut'].sort_values(ascending=False)

# Plotting the correlations with the target variable
plt.figure(figsize=(10, 8))
target_correlations.drop('clusterid - Prediction/OutPut').plot(kind='bar')  # Drop the target variable to avoid self-correlation
plt.title('Feature Correlation with Target Variable')
plt.xlabel('Features')
plt.ylabel('Correlation Coefficient')
plt.tight_layout()
plt.show()
```
