```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('tab_data.csv')

# Function to plot data
def plot_data(df, column, df_cleaned, is_categorical=False):
    if is_categorical:
        # For categorical data, use count plot
        fig, ax = plt.subplots(1, 2, figsize=(14, 5))
        sns.countplot(x=column, data=df, ax=ax[0])
        ax[0].set_title(f'Before Outlier Removal: {column}')
        ax[0].tick_params(axis='x', rotation=90)
        
        sns.countplot(x=column, data=df_cleaned, ax=ax[1])
        ax[1].set_title(f'After Outlier Removal: {column}')
        ax[1].tick_params(axis='x', rotation=90)
    else:
        # For numerical data, use histogram
        fig, ax = plt.subplots(1, 2, figsize=(14, 5))
        sns.histplot(df[column], kde=True, element='step', stat='density', ax=ax[0])
        ax[0].set_title(f'Before Outlier Removal: {column}')
        
        sns.histplot(df_cleaned[column], kde=True, element='step', stat='density', ax=ax[1])
        ax[1].set_title(f'After Outlier Removal: {column}')
    
    plt.tight_layout()
    plt.show()

# Identify numerical columns and handle outliers
numeric_cols = df.select_dtypes(include=[np.number]).columns
for column in numeric_cols:
    z_scores = np.abs(stats.zscore(df[column]))
    filter_outliers = z_scores < 3
    df_cleaned = df[filter_outliers]
    plot_data(df, column, df_cleaned)

# Handle categorical data
categorical_cols = ['companycode', 'clusterid - Prediction/OutPut']
for column in categorical_cols:
    plot_data(df, column, df, is_categorical=True)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('tab_data.csv')

# Function to plot data
def plot_data(df, column, df_cleaned):
    # For numerical data, use boxplot and strip plot
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    
    # Before cleaning
    sns.boxplot(y=df[column], ax=ax[0], color='lightblue', fliersize=0)
    sns.stripplot(y=df[column], ax=ax[0], color='darkblue', jitter=True, size=2)
    ax[0].set_title(f'Before Outlier Removal: {column}')
    ax[0].set_ylabel('Value')
    
    # After cleaning
    sns.boxplot(y=df_cleaned[column], ax=ax[1], color='lightgreen', fliersize=0)
    sns.stripplot(y=df_cleaned[column], ax=ax[1], color='darkgreen', jitter=True, size=2)
    ax[1].set_title(f'After Outlier Removal: {column}')
    ax[1].set_ylabel('Value')
    
    plt.tight_layout()
    plt.show()

# Identify numerical columns and handle outliers
numeric_cols = df.select_dtypes(include=[np.number]).columns
df_cleaned = df.copy()  # Initialize df_cleaned

for column in numeric_cols:
    if column not in ['companycode', 'clusterid - Prediction/OutPut']:
        z_scores = np.abs(stats.zscore(df[column]))
        filter_outliers = z_scores < 3
        df_cleaned = df_cleaned[filter_outliers]
        plot_data(df, column, df_cleaned[filter_outliers])
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('tab_data.csv')

# Function to plot data
def plot_data(df, column, df_cleaned):
    # Calculate z-scores for the column
    z_scores = np.abs(stats.zscore(df[column]))
    outliers = z_scores > 3

    # For numerical data, use histogram and scatter for outliers
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    
    # Before cleaning - histogram with outliers marked
    sns.histplot(df[column], kde=True, element='step', stat='density', ax=ax[0], color='lightblue')
    ax[0].scatter(df[column][outliers], np.zeros(sum(outliers)), color='red', zorder=2, label='Outliers')
    ax[0].set_title(f'Before Outlier Removal: {column}')
    ax[0].set_xlabel('Value')
    ax[0].set_ylabel('Density')
    ax[0].legend()
    
    # After cleaning - histogram
    sns.histplot(df_cleaned[column], kde=True, element='step', stat='density', ax=ax[1], color='lightgreen')
    ax[1].set_title(f'After Outlier Removal: {column}')
    ax[1].set_xlabel('Value')
    ax[1].set_ylabel('Density')
    
    # Set x-axis limits based on the data range for better scaling
    ax[0].set_xlim(df[column].min(), df[column].max())
    ax[1].set_xlim(df_cleaned[column].min(), df_cleaned[column].max())
    
    plt.tight_layout()
    plt.show()

# Identify numerical columns and handle outliers
numeric_cols = ['ff_assets_curr', 'ff_curr_ratio', 'ff_debt', 'ff_debt_eq', 'ff_ebitda_oper', 'ff_inven_turn', 'ff_invest_cf', 'ff_ltd_com_eq', 'ff_oper_cf', 'ff_roce', 'ff_sales']
df_cleaned = df.copy()  # Initialize df_cleaned

for column in numeric_cols:
    z_scores = np.abs(stats.zscore(df[column]))
    filter_outliers = z_scores < 3
    df_cleaned = df_cleaned[filter_outliers]
    plot_data(df, column, df_cleaned[filter_outliers])
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('tab_data.csv')

# Function to plot data
def plot_data(df, column, df_cleaned):
    # Calculate z-scores for the column
    z_scores = np.abs(stats.zscore(df[column]))
    outliers = z_scores > 3

    # For numerical data, use histogram and scatter for outliers
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    
    # Before cleaning - histogram with outliers marked
    sns.histplot(df[column], kde=True, element='step', stat='density', ax=ax[0], color='lightblue', bins=30)
    ax[0].scatter(df[column][outliers], np.zeros(sum(outliers)), color='red', zorder=2, label='Outliers')
    ax[0].set_title(f'Before Outlier Removal: {column}')
    ax[0].set_xlabel('Value')
    ax[0].set_ylabel('Density')
    ax[0].legend()
    
    # After cleaning - histogram
    sns.histplot(df_cleaned[column], kde=True, element='step', stat='density', ax=ax[1], color='lightgreen', bins=30)
    ax[1].set_title(f'After Outlier Removal: {column}')
    ax[1].set_xlabel('Value')
    ax[1].set_ylabel('Density')
    
    # Set x-axis limits based on the data range for better scaling
    ax[0].set_xlim(df[column].min(), df[column].max())
    ax[1].set_xlim(df_cleaned[column].quantile(0.01), df_cleaned[column].quantile(0.99))  # Zoom into the main distribution
    
    plt.tight_layout()
    plt.show()

# Identify numerical columns and handle outliers
numeric_cols = ['ff_assets_curr', 'ff_inven_turn', 'ff_curr_ratio', 'ff_debt', 'ff_invest_cf', 'ff_ltd_com_eq', 'ff_debt_eq', 'ff_oper_cf', 'ff_ebitda_oper', 'ff_roce', 'ff_sales']
df_cleaned = df.copy()  # Initialize df_cleaned

for column in numeric_cols:
    z_scores = np.abs(stats.zscore(df[column]))
    filter_outliers = z_scores < 3
    df_cleaned = df_cleaned[filter_outliers]
    plot_data(df, column, df_cleaned[filter_outliers])
```
